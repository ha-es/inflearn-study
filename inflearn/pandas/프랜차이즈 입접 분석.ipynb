```python
# 지도의 중심을 지정하기 위해 
# 위도와 경도의 평균 구하기

lat = df_31['위도'].mean()
long = df_31['경도'].mean()

lat, long
```

```python
m = folium.Map(location=[lat,long])
m
```

```python
# sample()값은 항상 랜덤
# 만약 항상 같은 값으로 갖고 오고 싶다면 random_state 사용
df_31.sample(random_state=31)
```

```python
# marker 찍기
folium.Marker([37.619059,126.915821],
              popup='<i>dounkin</i>',
              tooltip='던킨도너츠').add_to(m)

m.save('index.html') #한글 깨지는경우 html저장해서 확인
m
```

### 배스킨라빈스와 던킨도너츠 매장 분포

```python
df_31.index
```

```python
m = folium.Map(location=[lat,long], zoom_start=12)


for i in df_31.index:
        sub_lat = df_31.loc[i, '위도']
        sub_long = df_31.loc[i, '경도']
        title = df_31.loc[i, '상호명'] + "_" +  df_31.loc[i, '도로명주소']
        
        icon_color = "blue"
        
        if df_31.loc[i, '브랜드명'] == "던킨도너츠" :
            icon_color = "red"
        
        
    
        folium.Marker(        
            [sub_lat,sub_long],
            icon=folium.Icon(color=icon_color),
                popup=f'<i>{title}</i>',
                tooltip=title).add_to(m)

        
    
m.save('index.html') #한글 깨지는경우 html저장해서 확인
m
```

#### MarkerCluster 맵 그리기

```python
from folium.plugins import MarkerCluster


m = folium.Map(location=[lat,long], zoom_start=12)
#추가된 코드
marker_cluster = MarkerCluster().add_to(m)


for i in df_31.index:
        sub_lat = df_31.loc[i, '위도']
        sub_long = df_31.loc[i, '경도']
        title = df_31.loc[i, '상호명'] + "_" +  df_31.loc[i, '도로명주소']
        
        icon_color = "blue"
        
        if df_31.loc[i, '브랜드명'] == "던킨도너츠" :
            icon_color = "red"
        
        
    
        folium.Marker(        
            [sub_lat,sub_long],
            icon=folium.Icon(color=icon_color),
                popup=f'<i>{title}</i>',
                tooltip=title).add_to(marker_cluster)

        
    
m.save('index.html') #한글 깨지는경우 html저장해서 확인
m
```
